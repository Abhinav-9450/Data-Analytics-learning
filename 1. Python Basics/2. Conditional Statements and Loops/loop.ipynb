```python
#Que.01 sum of first n natural number using for and while loop
n = int(input("Enter a number: "))

#for loop
sum = 0

for i in range(1,n+1,1):
  sum += i
print(f"sum from for loop is {sum}")

#while loop
sum = 0
i = 1
while(i<=n):
  sum += i
  i = i+1
print(f"sum from while loop is {sum}")  
```
